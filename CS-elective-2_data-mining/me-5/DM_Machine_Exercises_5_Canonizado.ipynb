```python
# ==============================
# DENSITY USING FIXED RADIUS (ε-NEIGHBORHOOD)
# ==============================

import numpy as np
import matplotlib.pyplot as plt

# ==============================
# STEP 0: Define Points
# ==============================
# Each point represents a business location
# Columns: [X-coordinate, Y-coordinate]
# A simple 2D example for density computation

points = {
    "A": np.array([50, 5]),
    "B": np.array([55, 6]),
    "C": np.array([52, 5]),
    "D": np.array([100, 15]),
    "E": np.array([48, 4])
}

# Choose center point (the one whose neighborhood we analyze)
center_label = "A"
center = points[center_label]

print("Center point:", center_label, "at coordinates", center)
print("\n--- STEP 0 Complete ---\n")

# ==============================
# STEP 1: Choose radius
# ==============================
# Radius determines the neighborhood distance
# Larger radius → more points included → lower density if area grows faster
# Smaller radius → fewer points → higher density if area is small

r = 3 # You can experiment with other values (e.g., 3, 10)
print(f"Radius (r) = {r}")
print("\n--- STEP 1 Complete ---\n")

# ==============================
# STEP 2: Compute Euclidean distances from center
# ==============================
# Euclidean distance formula in 2D:
# d(A, B) = sqrt((x2 - x1)^2 + (y2 - y1)^2)

inside_points = []  # list to store points inside radius

print("Euclidean Distances from", center_label, ":\n")

for label, point in points.items():
    distance = np.linalg.norm(center - point)  # compute Euclidean distance
    print(f"Distance {center_label} → {label} = {distance:.2f}")

    # Check if the point is inside radius
    if distance <= r:
        inside_points.append(label)

print("\nPoints inside radius:", inside_points)
print("--- STEP 2 Complete ---\n")

# ==============================
# STEP 3: Count points inside radius
# ==============================
# Include the center itself in the count

total_inside = len(inside_points)
print("Total points inside radius =", total_inside)
print("--- STEP 3 Complete ---\n")

# ==============================
# STEP 4: Compute Area of Neighborhood (2D Circle)
# ==============================
# Area = π * r^2
# Used to compute density

area = np.pi * r**2
print("Area of neighborhood =", round(area, 2))
print("--- STEP 4 Complete ---\n")

# ==============================
# STEP 5: Compute Density
# ==============================
# Density = number of points / area
# Higher density → more points packed in same space

density = total_inside / area
print("Density(", center_label, ") =", round(density, 4))
print("--- STEP 5 Complete ---\n")

# ==============================
# STEP 6: Visualization
# ==============================
# Plot points and radius circle
plt.figure(figsize=(6,6))

for label, point in points.items():
    if label in inside_points:
        plt.scatter(point[0], point[1], color='blue', s=100, label='Inside' if label==inside_points[0] else "")
    else:
        plt.scatter(point[0], point[1], color='red', s=100, label='Outside' if label=='D' else "")

    plt.text(point[0]+0.5, point[1]+0.2, label)

# Draw circle for neighborhood
circle = plt.Circle(center, r, fill=False, color='green', linestyle='--', linewidth=2)
plt.gca().add_patch(circle)

# Formatting
plt.xlabel("X-coordinate")
plt.ylabel("Y-coordinate")
plt.title(f"ε-Neighborhood Density (r = {r})")
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# ==============================
# DENSITY USING FIXED RADIUS (ε-NEIGHBORHOOD) FOR ALL POINTS
# ==============================

import numpy as np
import matplotlib.pyplot as plt

# ==============================
# STEP 0: Define Points
# ==============================
points = {
    "A": np.array([50, 5]),
    "B": np.array([55, 6]),
    "C": np.array([52, 5]),
    "D": np.array([100, 15]),
    "E": np.array([48, 4])
}

labels = list(points.keys())

# ==============================
# STEP 1: Choose radius
# ==============================
r = 6
print(f"Radius (r) = {r}\n")

# ==============================
# STEP 2: Compute density for ALL points
# ==============================
densities = {}
clusters = {}

area = np.pi * r**2

for center_label, center_point in points.items():

    inside_points = []

    for label, point in points.items():
        distance = np.linalg.norm(center_point - point)

        if distance <= r:
            inside_points.append(label)

    total_inside = len(inside_points)
    density = total_inside / area

    densities[center_label] = density
    clusters[center_label] = inside_points

# ==============================
# STEP 3: Find densest point
# ==============================
densest_point = max(densities, key=densities.get)
densest_density = densities[densest_point]

print("Density for each point:\n")
for p in densities:
    print(f"{p}: {densities[p]:.4f}  | Neighborhood: {clusters[p]}")

print("\nDensest point:", densest_point)
print("Density value:", round(densest_density, 4))
print("Cluster around densest point:", clusters[densest_point])

# ==============================
# STEP 4: Visualization
# ==============================
plt.figure(figsize=(6,6))

center = points[densest_point]

for label, point in points.items():

    if label in clusters[densest_point]:
        plt.scatter(point[0], point[1], color='blue', s=120,
                    label="Cluster Member" if label == clusters[densest_point][0] else "")
    else:
        plt.scatter(point[0], point[1], color='red', s=120,
                    label="Outside Cluster" if label == "D" else "")

    plt.text(point[0]+0.5, point[1]+0.2, label)

# Draw circle of densest neighborhood
circle = plt.Circle(center, r, fill=False, color='green', linestyle='--', linewidth=2)
plt.gca().add_patch(circle)

plt.title(f"Densest Business Cluster (Center = {densest_point})")
plt.xlabel("X-coordinate")
plt.ylabel("Y-coordinate")
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```
