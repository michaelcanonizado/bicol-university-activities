```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from an online source
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
data = pd.read_csv(url)

# Display first five rows
print(data.head())

# Display dataset information
print(data.info())

# Display summary statistics
print(data.describe())

# Show unique values in the 'day' column
print(data['day'].unique())

# Count frequency of each day
print(data['day'].value_counts())

# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values
clean_data = data.dropna()
print(clean_data.shape)

# Average total bill by day
avg_bill = data.groupby('day')['total_bill'].mean()
print(avg_bill)

# Average tip by gender
avg_tip_gender = data.groupby('sex')['tip'].mean()
print(avg_tip_gender)

# Histogram of total bill
plt.hist(data['total_bill'])
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()

# Create a new column indicating high tips
data['high_tip'] = data['tip'].apply(lambda x: 1 if x > 5 else 0)

print(data[['tip', 'high_tip']].head())
```
